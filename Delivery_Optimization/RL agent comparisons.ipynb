# Environment definition
As for any RL problem, an environment must be designed where the Agent will evolve.<br>
For convenience and readability, we will define it like OpenAI Gym environments

```python
env = DeliveryEnvironment(n_stops = 10)

for i in [0,1,2,3]:
    env.step(i)
env.render()
```

# Q Learning
## Simple case

```python
env1 = DeliveryEnvironment(n_stops = 50,method = "distance")
env2 = DeliveryEnvironment(n_stops = 50,method = "distance")
agent2 = DeliveryQAgent(env1.observation_space,env1.action_space)
agent1 = QAgent(nA=env2.action_space, nS=env2.observation_space)


run_n_episodes(env1,agent1,"training_50_stops1.gif")
run_n_episodes(env2,agent2,"training_50_stops2.gif")
env1.render()
env2.render()
print('done')
```

## Traffic case

```python
env1 = DeliveryEnvironment(n_stops = 100,method = "traffic_box",box_size = 0.6,traffic_intensity = 100)
env2 = DeliveryEnvironment(n_stops = 100,method = "traffic_box",box_size = 0.6,traffic_intensity = 100)
agent2 = DeliveryQAgent(env1.observation_space,env1.action_space)
agent1 = QAgent(nA=env1.action_space, nS=env2.observation_space)

run_n_episodes(env1,agent1,"training_100_stops_traffic1.gif")
run_n_episodes(env2,agent2,"training_100_stops_traffic2.gif")
env1.render()
env2.render()

#run_n_episodes(env1,agent1,"training_100_stops_traffic1.gif")
#env1.render()
print('done')
```
